```python
import pandas as pd
from collections import Counter
```

```python
files =['KDD','CVPR_a','ACM','CVPR','ACM2','IEEE','IND','IND2','NIPS','ACM3']
```

```python
final_df = pd.DataFrame()
for f in files:
    df = pd.read_csv('../Data/AuthorDetails_Gr2021_'+f+'.csv',header=None)
    df.columns = ['Author_Name','Affiliations','Email','Conf']
    df['Conf'] = f
    final_df = final_df.append(df)
```

```python
final_df = final_df[final_df['Email'].notna()]
```

```python
final_df_no_dup = final_df.drop_duplicates(subset=['Email'])
final_df_no_dup.Conf.replace(to_replace={'CVPR_a': 'CVPR', 'ACM2': 'ACM', 'ACM3': 'ACM','IND2':'IND'}, inplace=True)
final_df_no_dup.reset_index(drop=True,inplace=True)
final_df_no_dup
```

```python
print(Counter(list(final_df_no_dup['Conf'])))
```

```python
final_df_no_dup.to_excel('../Data/scopus_authorList_2021_prev5Years_15_oct_2021.xlsx')
```

## Getting Maths and Pure science

```python
files_name =['maths','physics','chem','bio']
```

```python
final_df = pd.DataFrame()
for f in files_name:
    df = pd.read_csv('../Data/top_conf_micorosft_'+f+'.csv',error_bad_lines=False)
    df['Conf'] = f
    df = df.head(n=15)
    final_df = final_df.append(df)
```

```python
final_df = final_df.drop_duplicates(subset=['Name'])
```

```python
for f in files_name:
    print(f)
    df = final_df[final_df['Conf']==f]
    confList = list(set(df['Name']))
    str_ = ''
    for f in confList:
        str_+='CONF('+f+') OR '
    print(str_)
    print("\n\n+++++++++++++++++++++++\n\n")
```
